# 1. Basic population model
\begin{equation*}
N(t) = N_0 R^t. 
\end{equation*}


```python
%matplotlib inline   

from numpy import *
from pylab import *
mpl.rcParams['text.usetex'] = True
```

1 function in 1 figure

```python
R = 2
t = linspace(0.0, 10, num=60, endpoint=True) # create linear space from 0 to 10 which consists of 60 points including the end point
N = 1*R**t
plot(t,N,label=r"$N_0 = 1$") 
xlabel('Time $t$')    # label for x-axis
ylabel('No. of population $N$')    # label for y-axis
title('$N = N_0 R^t$ where $N_0 = 1$ and $R = 2$')     # title of figure
```

Several functions in 1 figure.

```python
R = 2
t = linspace(0.0, 10, num=60, endpoint=True) # create linear space from 0 to 10 which consists of 60 points including the end point
N1 = 1*R**t
N2 = 4*R**t
N3 = 8*R**t
plot(t,N1,label=r"$N_0 = 1$")
plot(t,N2,label=r"$N_0 = 4$")
plot(t,N3,label=r"$N_0 = 8$")  
legend(loc='upper center')   # generate the box that consists of lines and their label
xlabel('Time $t$')    # label for x-axis
ylabel('No. of population $N$')    # label for y-axis
title('$N = N_0 R^t$ where $N_0$ vary')     # title of figure
```
