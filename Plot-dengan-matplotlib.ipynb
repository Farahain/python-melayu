Mari kita mulakan dengan memplot graf 2D yang paling ringkas.

```python
# kod di bawah ini khusus untuk plot menggunakan iPython notebook supaya graf yang diplot dipaparkan di dalamnya.
%matplotlib inline   

from numpy import *
from matplotlib.pyplot import *

x = linspace(0.0, 2*pi, num=30, endpoint=True) # menghasilkan linear space dari 0 hingga 2*pi yang terdiri dari 30 titik termasuk titik terakhir pada 2*pi
y = sin(x)
plot(x,y)
```

Berjaya! Mudah sahaja kan? 
Apa kata kita cuba plot 2 fungsi sekaligus.

```python
x = linspace(0.0, 2*pi, num=30, endpoint=True) 
y1 = sin(x)
y2 = cos(x)
plot(x,y1,x,y2)
```

Alamak! Yang mana kos, yang mana sin ni? 
Mari kita cuba plot sekali lagi dan minta tolong Python untuk labelkan.

```python
x = linspace(0.0, 2*pi, num=30, endpoint=True) 
y1 = sin(x)
y2 = cos(x)
plot(x,y1,label='Fungsi sin x')
plot(x,y2,label='Fungsi cos x')
legend(loc='upper center')     # ini arahan minta tolong Python untuk buatkan kotak yang labelkan fungsi diplot 
```

Baiklah. Sekarang fungsi-fungsi diplot sudahpun dilabelkan.
Tapi kotak tu bertindih pula dengan garisan biru tu.
Mari kita cuba beri arahan kepada Python supaya besarkan ruang atas tu supaya tak bertindih.
Kemudian kita arahkan Python untuk tuliskan juga label pada paksi x dan y dan juga tajuk graf ini.

```python
plot(x,y1,label='Fungsi sin x')
plot(x,y2,label='Fungsi cos x')
legend(loc='upper right') 
axis([0, 2*pi, -2, 2])     # domain plot [mula x, hujung x, mula y, hujung y]
xlabel('$x$')    # label x dalam simbol berdasarkan LaTeX
ylabel('$y$')    # label y dalam simbol berdasarkan LaTeX
title('Belajar plot graf 2D')     # tajuk graf
```

Selain plot kedua-dua fungsi dalam satu plot, kita boleh juga asingkan menjadi 2 subplot tetapi masih dalam satu plot. Macam mana tu? Mari kita cuba subplotkan.

```python
subplot(2, 1, 1)
plot(x, y1)
title('Belajar Subplot')
ylabel(r'$\sin$ $\theta$')

subplot(2, 1, 2)
plot(x, y2)
xlabel(r'$ \theta $')
ylabel(r'$ \cos$ $\theta $')

```

Baiklah perkara terakhir untuk pelajaran kali ini ialah bagaimana untuk kita plot 3D.

```python
from numpy import *
from matplotlib.pyplot import *
from skimage import measure
from mpl_toolkits.mplot3d import Axes3D


def fun(x, y, z):
    return (x**2 + (9.0/4)*y**2 + z**2 -1)**3 -(x**2)*(z**3) -(9.0/80)*(y**2)*(z**3)


x, y, z = mgrid[-1.5:1.5:31j, -1.5:1.5:31j, -1.5:1.5:31j]
vol = fun(x, y, z)
verts, faces = measure.marching_cubes(vol, 0, spacing=(0.1, 0.1, 0.1))

fig = figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(verts[:, 0], verts[:,1], faces, verts[:, 2],
                cmap='OrRd', lw=1)
ax.view_init(elev=10, azim=100)
show()
```
